Searcher functions try to find parameters of a given curve. They do this by finding the Mean Squared Error rate (MSE rate) of a curve with its current parameters and the MSE rate gives how close the current parameters are to the actual parameters. 

Bayes' Information Criterion (BIC)
BIC = n * ln(MSE) + K * ln(n)
k = number of params
n = number of data points

The higher the BIC, the worse the model

Akaike's Information Criterion (AIC)
AIC = n * ln(MSE) + k * 2

AICc = n * ln(MSE) + k * 2n/n-k-1

rule of thumb:
lowest IC is the best model
models that have no more than 2 differences are not significantly different
